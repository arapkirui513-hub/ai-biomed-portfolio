```python
import numpy as np

# Load clinic_day.csv as pure numeric data (skip the PatientID column)
data = np.genfromtxt(
    "clinic_day.csv",
    delimiter=",",
    skip_header=1,
    usecols=(1, 2, 3)  # Glucose, HeartRate, OxygenSat
)

print("Data shape:", data.shape)  # should be (4, 3)

means = np.mean(data, axis=0)
stds = np.std(data, axis=0)
mins = np.min(data, axis=0)
maxs = np.max(data, axis=0)

print("Means (Glucose, HR, O2):", means)
print("Stds  (Glucose, HR, O2):", stds)
print("Mins  (Glucose, HR, O2):", mins)
print("Maxs  (Glucose, HR, O2):", maxs)
```

```python
theta = np.radians(45)
R = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])
v = np.array([1, 0])
rotated = R @ v
print(rotated)  # ~ [0.707, 0.707]
```

```python
# 1. Rotate 90 degrees — should give [0, 1] (pointing straight up)
theta90 = np.radians(90)
R90 = np.array([[np.cos(theta90), -np.sin(theta90)],
                [np.sin(theta90),  np.cos(theta90)]])
print(R90 @ np.array([1, 0]))

# 2. Rotate twice — should be same as rotating 90 degrees once
print(R90 @ (R90 @ np.array([1, 0])))   # 180 degrees total -> [-1, 0]

# 3. What is the determinant of any rotation matrix?
print(np.linalg.det(R90))   # always 1.0
```

```python
M = np.arange(9).reshape(3, 3)
print(M)
print("Diagonal:", np.diag(M))
# np.diag works on any 2D array. For a square matrix
```

```python
M2 = np.arange(12).reshape(3, 4)
print(M2)
print("Diagonal:", np.diag(M2))
# For a non-square matrix, NumPy just takes the diagonal up to the shortest dimension
```

```python
A = np.array([[1, 2],
              [3, 4]])
B = np.array([[5, 6],
              [7, 8]])

print("Matrix product:\n", np.dot(A, B))
# or more idiomatic:
print("Matrix product:\n", A @ B)
# Dot products generalise to matrix multiplication
#  @ is the cleanest operator for matrix multiplication.
```

```python
columns = ["Glucose", "HeartRate", "OxygenSat"]

means = np.mean(data, axis=0)
stds = np.std(data, axis=0)
mins = np.min(data, axis=0)
maxs = np.max(data, axis=0)

for col, mean, std, min_, max_ in zip(columns, means, stds, mins, maxs):
    print(f"{col:10} | mean={mean:.2f}, std={std:.2f}, min={min_}, max={max_}")
# Instead of raw arrays, you can pair results with column names
```

```python
normed = (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))
print(normed)
# Instead of normalising one column, you can normalise all at once
```
