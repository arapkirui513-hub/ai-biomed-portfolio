## 1. Machine Learning
A system that learns patterns from examples rather than being explicitly programmed with rules. Instead of writing "if chest X-ray shows white patch in lower lobe → flag as pneumonia," you show the model thousands of labelled X-rays and it figures out the rules itself.
Memory hook: teaching by example, not by instruction manual.

## 2. Training Set
The portion of your data the model actually learns from — it sees these examples during fit(). In your chest X-ray project this will be ~70% of the NIH dataset images.
Clinical parallel: the case studies a medical student learns from.

## 3. Validation Set
Data the model never trains on, but that you use to tune decisions — like choosing between DenseNet and ResNet, or adjusting the learning rate. You look at validation performance repeatedly during development.
Critical rule: because you keep peeking at it, it can leak into your decisions. That's why you need a third set.

## 4. Test Set
Completely locked away until you are 100% finished making all decisions. You evaluate on it exactly once, to report your final honest performance. If you use it to make any decision, it is no longer a test set — it has become a second validation set.
Clinical parallel: the unseen future patient. Your model has never met them.


## 5. Overfitting
The model has memorised the training data — including its noise, quirks, and flukes — rather than learning the underlying signal. Symptom: training accuracy very high, validation accuracy much lower. In clinical AI this is especially dangerous because a model trained on one hospital's data often overfits to that hospital's scanner, patient demographics, and labelling style.

## 6. Underfitting
The opposite problem: the model is too simple to capture the real patterns in the data. Both training and validation accuracy are low. Usually means you need a more complex model, more features, or more training.


## 7. Accuracy
Of all predictions made, what fraction were correct? The trap: on an imbalanced dataset (e.g. 99% healthy, 1% sick), a model that always predicts "healthy" gets 99% accuracy and is clinically worthless. Never use accuracy alone in medicine.

## 8. Precision
Of all the cases your model flagged as positive, how many actually were? High precision means few false alarms. A confirmatory test before surgery needs high precision — you don't want to operate on healthy patients.
Formula: TP / (TP + FP)

## 9. Recall (= Sensitivity)
Of all the actual positive cases, how many did your model catch? High recall means few missed cases. A screening test (chest X-ray triage) needs high recall — missing a real disease is dangerous.
Formula: TP / (TP + FN)

## 10. Sensitivity
Same as recall — just the clinical term for it. Used constantly in medical literature. "The test has 94% sensitivity" means it catches 94% of true cases.

## 11. Specificity
Of all the truly healthy patients, how many did your model correctly clear? High specificity means few false positives. Specificity and sensitivity pull in opposite directions — lowering your threshold raises sensitivity but drops specificity.
Formula: TN / (TN + FP)


## 12. AUC-ROC
The ROC curve plots sensitivity vs. (1 − specificity) at every possible classification threshold. The AUC (Area Under the Curve) summarises this into a single number: 0.5 means random chance, 1.0 means perfect. It tells you how well your model ranks sick above healthy, regardless of which threshold you choose. This is why your chest X-ray project will report AUC rather than accuracy.


## 13. Loss Function
A mathematical measure of how wrong the model's predictions are on the training data. During training, the model adjusts its internal weights to minimise this number. For classification tasks, binary cross-entropy is typical. You never directly optimise accuracy because it isn't differentiable — you optimise loss, which acts as a proxy.
Think of it as the model's score on a penalty system — lower is better.

## 14. Epoch
One complete pass through the entire training dataset. If you train for 10 epochs, the model has seen every training example 10 times. Too few epochs → underfitting. Too many → overfitting. You use the validation loss curve to decide when to stop.

## 15. Hyperparameter
A setting you choose before training begins that controls how training happens — learning rate, number of epochs, batch size, model architecture. These are distinct from the model's internal parameters (weights), which are learned automatically. You tune hyperparameters using the validation set, which is exactly why the validation set exists.
